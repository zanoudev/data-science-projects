##3. Insights
###3.1 Mobile Device Usage Dataset Insights
3.1.1 (Bivariate Analysis, Numerical/Numerical)

**Insight**: The more apps installed a user has, the more time they spend using their device.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=behaviour_ds["Screen On Time (hours/day)"], y=behaviour_ds["Number of Apps Installed"], alpha=0.7)

plt.xlabel("Screen On Time (hours/day)")
plt.ylabel("Number of Apps Installed")
plt.title("Scatter Plot: Screen On Time vs. Number of Apps Installed")

plt.show()
```

This scatter plot shows a positive correlation between Screen On Time (hours/day) and Number of Apps Installed. The distinct clusters indicate that users with more installed apps tend to spend more time using their devices. This supports the insight that users with a higher number of apps tend to have longer screen time.

3.1.2 (Bivariate Analysis, Numerical/Numerical)

 **Insight**: Users who have a higher screen time will drain their batteries faster

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(x=behaviour_ds["Screen On Time (hours/day)"], y=behaviour_ds["Battery Drain (mAh/day)"])
plt.xlabel("Screen on Time")
plt.ylabel("Battery Drain (mAh/day)")
plt.title("Battery Drain vs Screen on time")

# Show the plot
plt.show()
```

This scatter plot **Battery Drain vs. Screen On Time** shows a **positive correlation**, meaning that as **screen time increases, battery consumption (mAh/day) also increases**. The step-like clustering indicates distinct usage patterns, but overall, **users who spend more time on their screens experience higher battery drain**, supporting the insight that **higher screen time leads to faster battery depletion**.

3.2.3 (Bivariate Analysis, Numerical/Numerical)

Insight: Younger people tend to spend more time using their device

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(x=behaviour_ds["Age"], y=behaviour_ds["Screen On Time (hours/day)"],)

plt.xlabel("Age")
plt.ylabel("Screen Time")
plt.title("Bivariate Analysis: Age vs. Screen Time")
plt.show()
```

This scatter plot **Age vs. Screen Time** shows a higher density of points at **lower age values with higher screen time**, suggesting that **younger individuals tend to spend more time on their devices**. However, the distribution is spread across ages, indicating that screen time varies for all age groups, but is generally higher for younger users.

3.1.4 (Bivariate Analysis, Numerical/Numerical)

**Insight**: Users who have installed a lot of apps are more likely to have a higher data usage

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(x=behaviour_ds["Number of Apps Installed"], y=behaviour_ds["Data Usage (MB/day)"])
plt.xlabel("Number of Apps Installed")
plt.ylabel("Data Usage (MB/day)")
plt.title("Number of Apps Installed vs Data Usage (MB/day)")
plt.show()
```

This scatter plot **Number of Apps Installed vs. Data Usage (MB/day)** shows a **clear positive correlation**, where users with more installed apps tend to have **higher data usage**. The step-like clustering indicates distinct user behavior patterns, reinforcing the insight that **users with more apps are more likely to consume more mobile data daily**.

3.1.5 (Univariate analysis: Numerical)

**Insight:** Android devices are more common than iOS



```python
plt.figure(figsize=(12, 6))
category_counts = behaviour_ds["Operating System"].value_counts()
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("OS Distribution")
plt.show()
```

This pie chart **OS Distribution** visually represents that **Android devices (79.1%) are significantly more common than iOS devices (20.9%)**. This confirms the insight that **Android has a larger user base compared to iOS** in this dataset.

3.1.6 (Univariate analysis, Numerical)

Insight: Most users spend a 100 minutes or less using the app.

```python
sns.set(style="whitegrid")

plt.figure(figsize=(8, 5))
sns.histplot(behaviour_ds["App Usage Time (min/day)"], bins=20, kde=True)
plt.title("Distribution of App Usage Time (minutes/day)")
plt.xlabel("App Usage Time (min/day)")
plt.ylabel("Frequency")
plt.show()
```

This histogram **Distribution of App Usage Time (minutes/day)** shows that the **highest frequency of users** is concentrated around the **100 minutes or less** range. This supports the insight that **most users spend about 100 minutes or less using the app daily**, though there are some users with significantly higher usage times.

3.1.7 (Bivariate Analysis, Numerical/Categorical)

Insight: Females spend more time using their device.

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x=behaviour_ds["Gender"], y=behaviour_ds["Screen On Time (hours/day)"])
plt.title("Screen-On Time Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Screen On Time (hours/day)")
plt.show()
```

This **violin plot** of **Screen-On Time Distribution
by Gender** shows **similar distributions for both males and females**, with a slight indication that females may have a **higher median screen-on time**. However, the overall spread of usage is quite similar, suggesting that **both genders have comparable screen time habits**, though females might spend slightly more time on their devices on average.

3.1.8 (Univariate analysis, Numerical AND Bivariate analysis, Numerical/Numerical)

Insight: Older adults use the app the most.

```python
age_bins = [18, 25, 35, 45, 100]
age_labels = ["18-25", "26-35", "36-45", "46+"]
behaviour_ds["Age Group"] = pd.cut(behaviour_ds["Age"], bins=age_bins, labels=age_labels, right=False)

plt.figure(figsize=(7, 5))
sns.countplot(x=behaviour_ds["Age Group"], order=["18-25", "26-35", "36-45", "46+"])
plt.title("Age Group Distribution")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()


plt.figure(figsize=(8, 5))
sns.violinplot(x=behaviour_ds["Age Group"], y=behaviour_ds["Screen On Time (hours/day)"])
plt.title("Screen-On Time Distribution by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Screen On Time (hours/day)")
plt.show()
```

The **Age Group Distribution** bar chart shows that the **46+ age group has the highest number of users**, followed by the **26-35 and 36-45 age groups**, with **18-25 having the least users**.

However, the **Screen-On Time Distribution by Age Group** violin plot indicates that **screen-on time is fairly consistent across all age groups**, with no significant differences in median values. This suggests that **older adults have the highest representation in the dataset, but their individual screen time is not necessarily higher than younger groups**.

3.1.9 (Bivariate analysis, Numerical/Categorical)

Insight: Younger users consume more data than older users.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=behaviour_ds["Age Group"], y=behaviour_ds["Data Usage (MB/day)"], palette="muted")
plt.title("Data Usage Distribution by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Data Usage (MB/day)")
plt.show()
```

This **box plot of Data Usage Distribution by Age Group** does not strongly support the insight that younger users consume more data than older users. The **median data usage appears similar across all age groups**, with **some variability in the spread**. The **18-25** age group has a slightly higher median and interquartile range, but the **46+ group also shows significant data consumption**. This suggests that **data usage is relatively consistent across different age groups, rather than being dominated by younger users**.

3.1.10 (Bivariate Analysis, Numerical/Categorical)

Insight: iOS users have more apps installed than Android users.

```python
plt.figure(figsize=(7, 5))
sns.boxplot(x=behaviour_ds["Operating System"], y=behaviour_ds["Number of Apps Installed"], palette="muted")
plt.title("Number of Installed Apps by OS")
plt.xlabel("Operating System")
plt.ylabel("Number of Installed Apps")
plt.show()
```

This **box plot of Number of Installed Apps by OS** does not strongly support the insight that **iOS users have more apps installed than Android users**. The **medians and interquartile ranges appear very similar** for both operating systems. Both iOS and Android users have a wide spread of installed apps, with nearly identical maximum values. This suggests that the **number of installed apps is comparable between Android and iOS users**, rather than one group having significantly more.


###3.2 Panic Attack Dataset Insights
3.2.1. (Univariate analysis, categorical)

**Insight**: PTSD is the most common trigger for panic attacks

```python
category_counts = panic_ds["Trigger"].value_counts()
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.xlabel("Panic Attack Triggers")
plt.ylabel("Count")
plt.title("Distribution of Panic Attack Triggers")
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.grid(axis='y', linestyle="--", alpha=0.7)

plt.show()
```

This **bar chart of Panic Attack Triggers Distribution** does not support the insight that **PTSD is the most common trigger for panic attacks**. The counts for all triggers, including **PTSD, Phobia, Caffeine, Social Anxiety, Stress, and Unknown**, appear nearly **equal**, suggesting that **no single trigger dominates significantly over the others**. Therefore, it would be inaccurate to conclude that PTSD is the most common trigger based on this visualization.

3.2.2. (Bivariate Analysis, Numerical/Categorial)

**Insight**: PTSD and Social Anxiety show the highest panic scores

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=panic_ds["Trigger"], y=panic_ds["Panic_Score"])
plt.xlabel("Trigger")
plt.ylabel("Panic Score")
plt.title("Bivariate Analysis: Trigger vs. Panic Score")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle="--", alpha=0.6)
plt.show()
```

This **box plot of Trigger vs. Panic Score** does not strongly support the insight that **PTSD and Social Anxiety show the highest panic scores**. The median panic scores for **all triggers appear similar**, with **wide distributions across all categories**. PTSD and Social Anxiety do not have significantly higher medians or ranges compared to other triggers like **Stress, Phobia, or Caffeine**. Instead, the plot suggests that **panic scores are fairly distributed across different triggers**, rather than being notably higher for specific ones.

3.2.3 (Univariate Analysis, Numerical AND Bivariate Analysis, Numerical/Numerical)

Insight: Older patients experience panick attakcs more frequently.

```python
# MEAN VALUE
average_age = panic_ds["Age"].mean()
median_age = panic_ds["Age"].median()
print(f"Average Age: {average_age}")
print(f"Median Age: {median_age}")

age_bins = [18, 25, 35, 45, 100]
age_labels = ["18-25", "26-35", "36-45", "46+"]
panic_ds["Age Group"] = pd.cut(behaviour_ds["Age"], bins=age_bins, labels=age_labels, right=False)

plt.figure(figsize=(7, 5))
sns.countplot(x=panic_ds["Age Group"], order=["18-25", "26-35", "36-45", "46+"])
plt.title("Age Group Distribution")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()


plt.figure(figsize=(8, 5))
sns.violinplot(x=panic_ds["Age Group"], y=panic_ds["Panic_Attack_Frequency"])
plt.title("Panic Attacks Frequency Distribution by Age Group")
plt.xlabel("Age Group")
plt.ylabel("P.A Frequency")
plt.show()
```

The **violin plot of Panic Attack Frequency by Age Group** does not strongly support the insight that **older patients experience panic attacks more frequently**. The **median panic attack frequency appears similar across all age groups**, and the distributions show **a wide range of values for all groups**.

While the **46+ group has the highest representation in the dataset (as shown in the Age Group Distribution chart), their panic attack frequency does not appear significantly higher than younger groups**. Instead, panic attack frequency seems **consistent across different age groups**, suggesting that age is not a major determinant of panic attack frequency in this dataset.

3.2.4 (Bivariate Analysis, Numerical/Numerical)

Insight: Most people experience panic scores between 4 to 6 out of 10.


```python
category_counts = panic_ds["Panic_Score"].value_counts().sort_index()
print(category_counts)

plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.histplot(panic_ds["Panic_Score"], kde=True)
plt.xlabel("Panic Score")
plt.ylabel("Occurrences")
plt.title("Distribution of Panic Score")
plt.show()

ps_bins = [1, 4, 7, 10]
ps_labels = ["1-3", "4-6", "7+"]
panic_ds["Panic Score Group"] = pd.cut(panic_ds["Panic_Score"], bins=ps_bins, labels=ps_labels, right=False)

plt.figure(figsize=(7, 5))
sns.countplot(x=panic_ds["Panic Score Group"], order=["1-3", "4-6", "7+"])
plt.title("Panic Score Group Distribution (1 to 10)")
plt.xlabel("Panic Score Group")
plt.ylabel("Count")
plt.show()
```

The **histogram of Panic Score Distribution** and the **bar chart of Panic Score Group Distribution** support the insight that **most people experience panic scores between 4 to 6 out of 10**.

The bar chart shows that the **4-6 score group has the highest count**, followed by the **7+ group** and then the **1-3 group**. The histogram also highlights a peak in occurrences around **panic scores of 4 to 6**, reinforcing that **moderate panic scores are the most common in this dataset**.
